### ¿Qué grupo de tratamiento (trt) tiene mayor supervivencia después de 1000 días?


Para responder a la pregunta ¿Qué grupo de tratamiento (trt) tiene mayor supervivencia después de 1000 días? podemos hacer una __prueba t__ con las siguientes hipótesis:

    + __Ho:__ La media de supervivencia del grupo con tratamiento 1 es mayor que los del grupo con tratamiento 2
    + __Ha:__ La media de supervivencia del grupo con tratamiento 1 es menor o igual que los del grupo con tratamiento 2

Para responder a la pregunta ¿Los grupos de tratamiento tienen igual supervivencia después de 1000 días año? podemos hacer un __análisis de varianza__ con las siguientes hipótesis:

    + __Ho:__ No hay diferencia entre grupos: las medias para la variable dependiente (time, por ejemplo) son iguales
    + __Ha:__ Hay diferencias entre los dos grupos
    


Para esta pregunta la variable dependiente tiempo (time) es continua.

Si la variable dependiente fuera dicotómica (status), se convierte en un problema de clasificación.
    
Es importante determinar el tamaño de muestra cuando vamos a hacer el experimento (estudio). Cuando tenemos datos ya observados, debemos determinar si los datos son suficientes, así como la potencia del estudio con ese tamaño de muestra.

## ANOVA entre grupos

+ La variable dependiente (la variable de interés) debe ser una escala continua (es decir, los datos deben estar en una medición de intervalo o de proporción (razón)).
+ La variable independiente necesita tener dos grupos independientes con dos niveles. 
+ Cuando se prueban tres o más grupos categóricos independientes, es mejor usar un __ANOVA unidireccional o de una vía__. La prueba podría usarse para probar la diferencia entre solo dos grupos, sin embargo, una __prueba t__ de muestras independientes sería más apropiada.
- Los datos deben ser independientes de las observaciones (es decir, no debe haber los mismos participantes en ambos grupos).
+ La __variable dependiente debe tener una distribución normal o casi normal para cada grupo__. 
    
    + Vale la pena señalar que, si bien la prueba t es robusta para violaciones menores de la normalidad, si sus datos son muy anormales, valdría la pena usar una prueba no paramétrica o un bootstrapping.
+ No debe haber valores atípicos espurios.
+ Los datos deben tener homogeneidad de varianzas (__homocedasticidad__). Esta suposición se puede probar utilizando la prueba de homogeneidad de varianzas de Levene.


### Prueba t vs ANOVA

Cuando se trata de analizar la media de dos o más grupos de población, __ANOVA (Análisis de varianza) y la prueba t__ son las dos mejores prácticas preferidas, aunque hay una delgada línea de diferencia entre ambos.

La __prueba t se realiza cuando tienes que comparar las medias poblacionales entre dos grupos__. Pero __cuando hay tres o más grupos, se realiza la prueba ANOVA__.

Tanto la prueba t como ANOVA son métodos estadísticos para probar una hipótesis. Ambas comparten los supuestos:

- La muestra extraída de la población se distribuye normalmente
- Varianza homogénea (homocedasticidad
- Muestreo aleatorio de datos
- Las observaciones son independientes.
- La variable dependiente se mide en niveles de razón o de intervalo


__Las pruebas de hipótesis pueden ser de una cola o de dos colas__.

+ Si la hipótesis alternativa emplea __“>” o “<” se trata de un test de una cola__, en el que solo se analizan desviaciones en un sentido. 
    + Solo se emplean test de una cola cuando se sabe con seguridad que las desviaciones de interés son en un sentido y únicamente si se ha determinado antes de observar la muestra, no a posteriori.

+ __Si se la hipótesis alternativa es del tipo “diferente de” se trata de un test de dos colas__, en el que se analizan posibles __desviaciones en las dos direcciones__. 

## ANOVA de una via (unidireccional) o de un solo factor

Los supuestos puede encontrarlos detalladamente explicados en: https://statistics.laerd.com/spss-tutorials/one-way-anova-using-spss-statistics.php

__Suposición n.º 1:__ su variable dependiente debe medirse en el nivel de intervalo o razón (es decir, son continuos). Ejemplos de variables que cumplen con este criterio: tiempo de revisión de un proceso, la inteligencia medida mediante la puntuación de coeficiente intelectual, el rendimiento en el examen medido de 0 a 100, etc.

__Suposición n.º 2:__ su variable independiente debe consistir en dos o más grupos categóricos independientes. Por lo general, un ANOVA unidireccional se usa cuando tiene tres o más grupos categóricos e independientes, pero se puede usar solo para dos grupos (pero una prueba t de muestras independientes se usa más comúnmente para dos grupos). Ejemplos de variables independientes que cumplen con este criterio incluyen etnicidad (p. ej., 3 grupos: caucásico, afroamericano e hispano), nivel de actividad física (p. ej., 4 grupos: sedentario, bajo, moderado y alto), etc.

__Suposición n.º 3:__ debe tener independencia de las observaciones, lo que significa que no hay relación entre las observaciones de cada grupo o entre los grupos mismos. Por ejemplo, debe haber diferentes participantes en cada grupo sin que ningún participante esté en más de un grupo. Esto es más un problema de diseño del estudio que algo que pueda probar, pero es una suposición importante del ANOVA unidireccional. Si su estudio no cumple con esta suposición, deberá usar otra prueba estadística en lugar del ANOVA unidireccional (por ejemplo, un diseño de __medidas repetidas__). Si no está seguro de si su estudio cumple con esta suposición, puede utilizar el Selector de pruebas estadísticas, que es una buena guía https://statistics.laerd.com/features-selecting-tests.php

__Suposición n.º 4:__ No debe haber valores atípicos significativos. Busque las opciones que tiene para tratar con valores atípicos. 

__Suposición n.º 5:__ Su variable dependiente debe tener una __distribución aproximadamente normal__ para cada categoría de la variable independiente. Solo deben ser aproximadamente normales porque es bastante "robusto" a las violaciones de la normalidad, lo que significa que la suposición puede violarse un poco y aun así proporcionar resultados válidos. Puede probar la normalidad con la __prueba de normalidad de Shapiro-Wilk__. También puede aplicar transformaciones si sus datos no cumplen con esta suposición (es decir, si fallan más que un poco). 

__Suposición n.º 6:__ Tiene que haber homogeneidad de varianzas. Puede probar esta suposición utilizando la __prueba de homogeneidad de varianzas de Levene__. Si sus datos no cumplen con esta suposición, no solo necesitará realizar un ANOVA de Welch (no paramétrica; https://pythonfordatascienceorg.wordpress.com/welch-t-test-python-pandas/) en lugar de un ANOVA unidireccional, sino también usar una prueba post hoc diferente (prueba de Games-Howell, en lugar de una prueba post hoc de Tukey). 
Una prueba post-hoc se realiza despuresponde a la pregunta: ¿Qué grupos difieren significativamente de los demás en términos de la media?

https://usq.pressbooks.pub/statisticsforresearchstudents/chapter/one-way-anova-assumptions/


### ANOVA de medidas repetidas

Se utiliza un ANOVA de medidas repetidas para determinar si existe o no una diferencia estadísticamente significativa entre las medias de tres o más grupos en los que __aparecen los mismos sujetos en cada grupo__.

Tiene 3 supuestos. https://www.statology.org/repeated-measures-anova-assumptions/

1. Independencia: Cada una de las observaciones debe ser independiente.

2. Normalidad: La distribución de la variable de respuesta se distribuye normalmente.

3. Esfericidad (sphericity): Las varianzas de las diferencias entre todas las combinaciones de grupos relacionados deben ser iguales.

Si se viola una o más de estas suposiciones, entonces los resultados del ANOVA de medidas repetidas pueden no ser confiables.

El de esfericidad: https://www.statology.org/mauchlys-test-of-sphericity/

La prueba de esfericidad de Mauchly usa las siguientes hipótesis:

    H0: The variances of the differences are equal
    
    HA: The variances of the differences are not equal

https://www.geeksforgeeks.org/how-to-perform-a-repeated-measures-anova-in-python/

### Algunos elementos sobre el ANOVA de una vía en Python con f_oneway() de scipy

El ANOVA unidireccional o de una vía prueba la hipótesis nula de que dos o más grupos tienen la misma media poblacional. La prueba se aplica a muestras de dos o más grupos, posiblemente con diferentes tamaños.

+ Con f_oneway de scipy:
    + Si estas suposiciones no son ciertas para un conjunto dado de datos, aún puede ser posible utilizar la prueba H de Kruskal-Wallis (`scipy.stats.kruskal`) o la prueba Alexander-Govern (`scipy.stats.alexandergovern`) aunque con alguna pérdida de potencia.
    + La longitud de cada grupo debe ser de al menos uno, y debe haber al menos un grupo con longitud mayor que uno. Si estas condiciones no están satisfechos, se genera una advertencia y (``np.nan``, ``np.nan``) es regresado.
    + Si todos los valores de cada grupo son idénticos y existen al menos dos grupos con diferentes valores, la función genera una advertencia y devuelve (``np.inf``, 0).
    + Si todos los valores en todos los grupos son iguales, la función genera una advertencia y devuelve (``np.nan``, ``np.nan``).
    + El algoritmo es de Heiman [G.W. Heiman, "Understanding research methods and statistics: An integrated introduction for psychology", Houghton, Mifflin and Company, 2001.]_, pp.394-7.

## Potencia (power) de un estudio: 1- beta

Múltiples formas de interpretar la potencia correctamente:

+ La potencia es la probabilidad de rechazar la hipótesis nula cuando, en realidad, es falsa. ¡Eso es bueno: la rechazo y es falsa!
+ La potencia es la probabilidad de tomar una decisión correcta (rechazar la hipótesis nula) cuando la hipótesis nula es falsa.
+ La potencia es la probabilidad de que una prueba de significancia detecte un efecto que está presente.
+ La potencia es la probabilidad de que una prueba de significancia detecte una desviación de la hipótesis nula, en caso de que exista dicha desviación.
+ La potencia es la probabilidad de evitar un error de tipo II.

#### qqplot()

sm.qqplot() : Gráfico Q-Q de los cuantiles de x frente a los cuantiles/ppf de una distribución.

Es un diagrama de dispersión creado al graficar dos conjuntos de cuantiles uno contra el otro. Si ambos conjuntos de cuantiles provienen de la misma distribución, deberíamos ver los puntos formando una línea más o menos recta.

+ La función de punto de probabilidad o __ppf__ es la inversa de la __cdf (Cumulative density function)__. Específicamente, la ppf devuelve el punto exacto donde la probabilidad de todo a la izquierda es igual a un valor dado 'x' . Esto se puede considerar como la función percentil ya que la ppf nos dice el valor de un percentil dado de los datos. Si 'x' es 1- alpha (nivel de significancia; 0.95, por ejemplo)
+ cdf es a probabilidad de todos los resultados menores o iguales a un valor 'x' dado. Gráficamente, esta es el área total de todo lo que es menor o igual a 'x' (el área total a la izquierda de x)

### Cómo manejar datos no normales

Si un conjunto de datos dado no tiene una distribución normal, a menudo podemos realizar una de las siguientes transformaciones  para que tenga una distribución más normal:

1. Transformación logarítmica: transforma los valores de x a log(x).
2. Transformación de raíz cuadrada: transforma los valores de x a √x.
3. Transformación de raíz cúbica: transforma los valores de x a x1/3.
4. Estandarizaciones:

    4.1 Transformación de potencia: Box-Cox y Yeo_Johnson
    4.2 Transformación de cuantil

Al realizar estas transformaciones, el conjunto de datos suele tener una distribución más normal.

Box-Cox requiere que los datos de entrada sean estrictamente positivos, mientras que Yeo-Johnson admite datos tanto positivos como negativos. De forma predeterminada, la normalización de varianza unitaria de media cero se aplica a los datos transformados, por lo que no es necesario aplicar la estandarización.

Ver https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PowerTransformer.html

## Prueba t con scipy

Calcula la prueba T para las medias de *dos muestras independientes* de puntajes.

Esta es una prueba para la hipótesis nula de que 2 muestras independientes tienen valores promedio (esperados) idénticos. Esta prueba asume que las poblaciones tienen varianzas idénticas por defecto.

### de una cola
    + __Ho:__ La media de supervivencia del grupo con tratamiento 1 es mayor que los del grupo con tratamiento 2
    + __Ha:__ La media de supervivencia del grupo con tratamiento 1 es menor o igual que los del grupo con tratamiento 2

```python
from scipy import stats
```

```python
stats.ttest_ind?
```

```python
res = stats.ttest_ind(df1.cbr_time, df2.cbr_time, 
                      equal_var=True)

display(res)
```

```python
resless = stats.ttest_ind(df1.cbr_time, df2.cbr_time, alternative='less',
                      equal_var=True)

display(resless)
```

### Interpretación
- Se obtienen los mismos resultados que con el método de pingouin, pero la salida tiene menos información

```python
resless2 = stats.ttest_ind(df1.cbr_time, df2.cbr_time, alternative='less',permutations=10000,
...                 random_state=33, equal_var=True)

display(resless2)
```

```python
resless3 = stats.ttest_ind(df1.cbr_time, df2.cbr_time, alternative='greater',
                           random_state=33, equal_var=True,trim=0.4)
display(resless3)
```
