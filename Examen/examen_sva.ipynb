### 11.

Dada la siguiente función generadora que produce una contraseña al azar, seleccione todas las respuestas válidas:  

+ print ("Password is ", randomPassword()) imprime el password

+ X El ciclo siguiente imprime el password  
for i in randomPassword():
    print(i)


+ X El password tiene 8 caracteres

+ Si eliminamos la sentencia  
     random.SystemRandom().shuffle(passwordList)  
ya no obtenemos el password

```python
import random
import string
def randomPassword():
    randomSource = string.ascii_letters + string.digits + string.punctuation
    password = random.sample(randomSource, 4)
    password += random.sample(string.ascii_uppercase, 2)
    password += random.choice(string.digits)
    password += random.choice(string.punctuation)
    passwordList = list(password)
    random.SystemRandom().shuffle(passwordList)
    password = ''.join(passwordList)
    yield password

print ("Password is ", next(randomPassword()))
```
